# 1. şık cevabı

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
%matplotlib notebook

ellitekrar=50

ortalama1=[]
for i in range (1,ellitekrar+1):
    a= np.random.normal(32,8,1000)
    ortalama1.append(np.average(a))
                        
    
    
```

```python
bintekrar=1000

ortalama2=[]
for y in range (1,bintekrar+1):
    c= np.random.normal(45,9,1000)
    ortalama2.append(np.average(c))
```

```python
plt.figure(figsize=(10,5), dpi=100)

plt.subplot(1,2,1)
plt.hist(ortalama1)

plt.subplot(1,2,2)
plt.hist(ortalama2)

plt.show()
```

```python
örnek_veri_1= np.random.normal(np.mean(ortalama1),np.std(ortalama1),50)

örnek_veri_1.sort()
ortalama1.sort()

plt.figure(figsize=(10,5),dpi=100)
plt.subplot(1,2,1)
baslik_font = {"family":"arial", "color":"darkred", "weight":"bold", "size":13}
eksen_font = {"family":"arial", "color":"darkblue", "weight":"bold","size":12}
plt.title("NORMAL DAĞILIM 1& ORTALAMA 1", fontdict=baslik_font)
plt.xlabel("NORMAL DAĞILIM 1", fontdict=eksen_font)
plt.ylabel("ORTALAMA 1", fontdict=eksen_font)
plt.scatter(örnek_veri_1,ortalama1)
plt.plot([np.min(örnek_veri_1),np.max(örnek_veri_1)],[np.min(ortalama1),np.max(ortalama1)], c="r")

plt.subplot(1,2,2)
örnek_veri_2= np.random.normal(np.mean(ortalama2),np.std(ortalama2),1000)


plt.title("NORMAL DAĞILIM 2 & ORTALAMA 2", fontdict=baslik_font)
plt.xlabel("NORMAL DAĞILIM 2", fontdict=eksen_font)
plt.ylabel("ORTALAMA 2", fontdict=eksen_font)
plt.scatter(örnek_veri_2,ortalama2)
plt.plot([np.min(örnek_veri_2),np.max(örnek_veri_2)],[np.mean(ortalama2),np.max(ortalama2)], c="r")

plt.tight_layout()
plt.show()
```

# 2. şık cevabı

```python
adil=[]
adil_ortalama=[]

for atıs_sayisi in range (1,1000):
    atış = np.random.binomial(1,0.5)
    adil.append(atış)
    adil_ortalama.append(np.mean(atış))
    
```

```python
adil_olmayan=[]
adil_olmayan_ortalama=[]

for atış_sayısı2 in range(1,1000):
    atış2 = np.random.binomial(1,0.75)
    adil_olmayan.append(atış2)
    adil_olmayan_ortalama.append(np.mean(atış2))
```

```python
plt.figure(figsize=(10,4),dpi=100)
baslik_font = {"family":"arial", "color":"darkred","weight":"bold", "size":13}
eksen_font = {"family":"arial", "color":"darkblue","weight":"bold","size":10}

plt.subplot(1,2,1)
plt.plot(adil_ortalama[0:100])
plt.hlines(0.5, 0,100, colors= 'r' )
plt.title("ADİL OLANININ TURA GELME OLASILIĞI", fontdict=baslik_font)
plt.xlabel("ADİL ATIŞ",fontdict=eksen_font)
plt.ylabel("ADİL ATIŞ ORTALAMA",fontdict=eksen_font)

plt.subplot(1,2,2)
plt.plot(adil_olmayan_ortalama[0:100])
plt.hlines(0.75, 0,100, colors= 'r' )
plt.title("ADİL OLMAYANIN TURA GELME OLASILIĞI", fontdict=baslik_font)
plt.xlabel("ADİL OLMAYAN ATIŞ",fontdict=eksen_font)
plt.ylabel("ADİL OLMAYAN ATIŞ ORTALAMA",fontdict=eksen_font)

plt.tight_layout()
plt.show()
```
