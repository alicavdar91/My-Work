# 2. şık cevabı

```python
from scipy.stats import zscore

aşırı_deger = zscore(eğitim_sistemi2["TOTAL_REVENUE"])
for sayı in range(1,5):
    print("Eşik değeri = {}".format(sayı))
    print("Aykırı değerlerin sayısı: {}".format(len(np.where(aşırı_deger>sayı)[0])))
    print("-------------------")
```

```python
aşırı_deger2 = zscore(eğitim_sistemi2["TOTAL_EXPENDITURE"])
for sayı2 in range(1,5):
    print("Eşik değeri = {}".format(sayı2))
    print("Aykırı değerlerin sayısı: {}".format(len(np.where(aşırı_deger2>sayı2)[0])))
    print("-------------------")
```

```python
q75,q25 = np.percentile ( eğitim_sistemi2["TOTAL_EXPENDITURE"],[75,25])
caa = q75-q25
```

```python
eşik_değerleri= pd.DataFrame(columns=["'esik_degeri', 'aykiri_deger_sayisi'"])
for eşik_değeri in np.arange(1,5,1):
    min_deger = q25-(caa*eşik_değeri)
    max_deger = q75+(caa*eşik_değeri)
    aykırı_deger_sayisi = len (np.where((eğitim_sistemi2["TOTAL_EXPENDITURE"]>max_deger)|
                                        (eğitim_sistemi2["TOTAL_EXPENDITURE"]<min_deger ))[0])
    eşik_değerleri = eşik_değerleri.append({"EŞİK DEĞERİ":eşik_değeri, "AYKIRI DEĞER SAYISI":aykırı_deger_sayisi},
                                           ignore_index=True)
eşik_değerleri   
```

```python
from scipy.stats.mstats import winsorize
winsorize_toplam_gelir= winsorize(eğitim_sistemi2["TOTAL_REVENUE"],(0,0.10))
winsorize_toplam_gelir
```

```python
eğitim_sistemi2["winsorize_TOTAL_REVENUE"]=winsorize_toplam_gelir
```

```python
winsorize_toplam_harcama= winsorize(eğitim_sistemi2["TOTAL_EXPENDITURE"],(0,0.10))
winsorize_toplam_harcama
```

```python
eğitim_sistemi2["winsorize_TOTAL_EXPENDITURE"]=winsorize_toplam_harcama
```

```python
plt.boxplot(winsorize_toplam_gelir)
plt.title("WİNSORİZE TOPLAM GELİR KUTU GRAFİĞİ(whis=1.5)", fontdict=baslik_font)
plt.show()
```

```python
plt.boxplot(winsorize_toplam_harcama)
plt.title("WİNSORİZE TOPLAM HARCAMA KUTU GRAFİĞİ(whis=1.5)", fontdict=baslik_font)
plt.show()
```
