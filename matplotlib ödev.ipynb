```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd


import warnings
warnings.filterwarnings('ignore')
```

```python
kalp = pd.read_csv("kalp.csv")
kalp
```

```python

data=pd.DataFrame()
data["YAŞ ON"] = np.random.choice(kalp["age"], 15)
data["KOLESTEROL ON"] = np.random.choice(kalp["chol"],15)
y_pos_yas_on = np.arange(len(data["YAŞ ON"]))
y_pos_kolesterol_on= np.arange(len(data["KOLESTEROL ON"]))

bar_width = 0.7
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
birinci_veri = plt.bar(y_pos_yas_on,data["YAŞ ON"], bar_width,color="r")
plt.xlabel("YAŞ")
plt.ylabel("İNDEX")
plt.title ("YAŞIN KALP KRİZİNE ETKİSİ")

plt.subplot(1,2,2)
ikinci_veri = plt.bar(y_pos_kolesterol_on, data["KOLESTEROL ON"], bar_width,color="g")
plt.xlabel("KOLESTEROL")
plt.ylabel("İNDEX")
plt.title ("KOLESTEROLÜN KALP KRİZİNE ETKİSİ")
plt.show()
```
