```python
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
```

```python
performans["hedef_degiskeni"] = performans["math score"]+performans["reading score"]+performans["writing score"]
```

```python
plt.hist(performans["hedef_degiskeni"])
plt.show()
```

```python
winsorize_hedef_degiskeni=winsorize(performans["hedef_degiskeni"],(0,0.10))
```

```python
plt.hist(winsorize_hedef_degiskeni)
plt.show()
```
