```python
def null_deger_hesapla(df):
    yüzde=df.isnull().sum()*100/df.shape[0]
    return yüzde
```

```python
null_deger_hesapla(eğitim)
```

```python
def null_deger_hesaplaa(df):
    yüzde = df.isnull().sum()*100/df.shape[0]
    adet = df.isnull().sum()
    return pd.DataFrame(list(zip(adet,yüzde)),columns = ["adet","yüzde"], index=df.columns)
```

```python
null_deger_hesaplaa(eğitim)
```

# 3. şık cevabı

# 1.yöntem

```python
eğitim_list=["ENROLL","TOTAL_REVENUE","FEDERAL_REVENUE","STATE_REVENUE","LOCAL_REVENUE",
             "TOTAL_EXPENDITURE","INSTRUCTION_EXPENDITURE","GRADES_4_G","GRADES_8_G","GRADES_12_G",
             "GRADES_1_8_G","GRADES_9_12_G","GRADES_ALL_G","AVG_MATH_4_SCORE","AVG_MATH_8_SCORE",
             "AVG_READING_4_SCORE","AVG_READING_8_SCORE"]
```

```python
eğitim2 = eğitim.copy()
yıllar = eğitim2["YEAR"].unique()
for col in eğitim_list:
    for yıl in yıllar:
        YIL_SART = eğitim2["YEAR"] == yıl
        eğitim2.loc[YIL_SART, col] = eğitim2.loc[YIL_SART, col].fillna(eğitim2[YIL_SART][col].mean())
```

```python
null_deger_hesaplaa(eğitim2)
```

# 4. şık cevabı

```python
eğitim3 = eğitim.copy()

for col in eğitim_list:
    eğitim3[col] = eğitim3.groupby("YEAR")[col].transform(lambda x:x.interpolate())
eğitim3
```

```python
null_deger_hesaplaa(eğitim3)
```
