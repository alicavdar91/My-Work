```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats.mstats import winsorize
import seaborn as sns
import warnings
import matplotlib.pyplot as plt 

warnings.filterwarnings("ignore")
sns.set(style="whitegrid")
pd.options.display.float_format= "{:.3f}".format
plt.rcParams["figure.dpi"]=100
plt.rcParams["figure.figsize"]=(8,5.5)
```

# 1.ŞIK CEVAP

```python
öğrenci_performans = pd.read_csv("öğrenci_performans.csv")
öğrenci_performans.columns=['cinsiyet', 'koken', 'aile_egt_seviyesi', 'ogle_yemegi', 
                              'hazirlik_kursu', 'matematik_notu', 'okuma_notu', 'yazma_notu']
öğrenci_performans
```

```python
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 15 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 12 }
```

```python
plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
sns.barplot(öğrenci_performans["cinsiyet"],öğrenci_performans["matematik_notu"])
plt.title("CİNSİYET & MATEMETİK NOTU İLİŞKİSİ", fontdict=baslik_font)
plt.xlabel("CİNSİYET", fontdict=eksen_font)
plt.ylabel("MATEMATİK NOTU", fontdict=eksen_font)

plt.subplot(1,3,2)
sns.barplot(öğrenci_performans["cinsiyet"],öğrenci_performans["okuma_notu"])
plt.title("CİNSİYET & OKUMA NOTU İLİŞKİSİ", fontdict=baslik_font)
plt.xlabel("CİNSİYET", fontdict=eksen_font)
plt.ylabel("OKUMA NOTU", fontdict=eksen_font)

plt.subplot(1,3,3)
sns.barplot(öğrenci_performans["cinsiyet"],öğrenci_performans["yazma_notu"])
plt.title("CİNSİYET & YAZMA NOTU İLİŞKİSİ", fontdict=baslik_font)
plt.xlabel("CİNSİYET", fontdict=eksen_font)
plt.ylabel("YAZMA NOTU", fontdict=eksen_font)

plt.show()
```

```python
plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
sns.barplot(öğrenci_performans["koken"],öğrenci_performans["matematik_notu"])
plt.title("KÖKEN & MATEMETİK NOTU İLİŞKİSİ", fontdict=baslik_font)
plt.xlabel("KÖKEN", fontdict=eksen_font)
plt.ylabel("MATEMATİK NOTU", fontdict=eksen_font)

plt.subplot(1,3,2)
sns.barplot(öğrenci_performans["koken"],öğrenci_performans["okuma_notu"])
plt.title("KÖKEN & OKUMA NOTU İLİŞKİSİ", fontdict=baslik_font)
plt.xlabel("KÖKEN", fontdict=eksen_font)
plt.ylabel("OKUMA NOTU", fontdict=eksen_font)

plt.subplot(1,3,3)
sns.barplot(öğrenci_performans["koken"],öğrenci_performans["yazma_notu"])
plt.title("KÖKEN & YAZMA NOTU İLİŞKİSİ", fontdict=baslik_font)
plt.xlabel("KÖKEN", fontdict=eksen_font)
plt.ylabel("YAZMA NOTU", fontdict=eksen_font)

plt.show()
```

```python
plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
sns.barplot(öğrenci_performans["aile_egt_seviyesi"],öğrenci_performans["matematik_notu"])
plt.title("AİLENİN EĞİTİM SEVİYESİ & MATEMETİK NOTU İLİŞKİSİ", fontdict=baslik_font)
plt.xlabel("AİLENİN EĞİTİM SEVİYESİ", fontdict=eksen_font)
plt.ylabel("MATEMATİK NOTU", fontdict=eksen_font)

plt.subplot(1,3,2)
sns.barplot(öğrenci_performans["aile_egt_seviyesi"],öğrenci_performans["okuma_notu"])
plt.title("AİLENİN EĞİTİM SEVİYESİ & OKUMA NOTU İLİŞKİSİ", fontdict=baslik_font)
plt.xlabel("AİLENİN EĞİTİM SEVİYESİ", fontdict=eksen_font)
plt.ylabel("OKUMA NOTU", fontdict=eksen_font)

plt.subplot(1,3,3)
sns.barplot(öğrenci_performans["aile_egt_seviyesi"],öğrenci_performans["yazma_notu"])
plt.title("AİLENİN EĞİTİM SEVİYESİ & YAZMA NOTU İLİŞKİSİ", fontdict=baslik_font)
plt.xlabel("AİLENİN EĞİTİM SEVİYESİ", fontdict=eksen_font)
plt.ylabel("YAZMA NOTU", fontdict=eksen_font)

plt.tight_layout()
plt.show()
```
