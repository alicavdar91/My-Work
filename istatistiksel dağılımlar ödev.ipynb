```python
import yfinance as yf
```

```python
google= yf.download("GOOG", start ="2015-01-01", end="2017-01-01")
```

```python
google
```

```python
getiri = []

n=0
while n<504:
    getiri.append((google["Open"][n+1]-google["Open"][n])/google["Open"][n])
    n+=1
```

```python
normal_dağlım_getiri= np.random.normal(np.mean(getiri),np.std(getiri),504)
```

```python
getiri.sort()
normal_dağlım_getiri.sort()

plt.figure(figsize=(8,4),dpi=100)
eksen_font={"family":"arial","color":"darkblue","weight":"bold","size":10}
baslık_font={"family":"arial","color":"darkred","weight":"bold","size":13}
plt.title("GOOGLE HİSSE SENETİ & NORMAL DAĞILIM VERİSİ", fontdict=baslık_font)
plt.xlabel("NORMAL DAĞILIM VERİSİ",fontdict=eksen_font)
plt.ylabel("GOOGLE HİSSE SENETİ", fontdict=eksen_font)
plt.scatter(getiri,normal_dağlım_getiri)
plt.plot([np.min(getiri),np.max(getiri)],[np.min(normal_dağlım_getiri),np.max(normal_dağlım_getiri)],c="r")
plt.show()
```
